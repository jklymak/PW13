```python
def getDist(cgrid):
    import seawater as sw
    x=0.*cgrid['longitude']
    x[0]=0.
    for i in arange(0,shape(cgrid['temp'])[1]-1):
        dx,head = sw.dist(cgrid['longitude'][[i,i+1]],cgrid['latitude'][[i,i+1]],units='km')
        x[i+1]=x[i]+dx
    return x
        
import numpy as np

def pointlinedist(x1,y1, x2,y2, x3,y3): # x3,y3 is the point
    px = x2-x1
    py = y2-y1

    something = px*px + py*py

    u =  ((x3 - x1) * px + (y3 - y1) * py) / float(something)

    u[u>1.]=1.
    u[u<0.]=0.
    
    x = x1 + u * px
    y = y1 + u * py

    dx = x - x3
    dy = y - y3

    # Note: If the actual distance does not matter,
    # if you only want to compare what this function
    # returns to other results of this function, you
    # can just return the squared distance instead
    # (i.e. remove the sqrt) to gain a little performance

    dist = np.sqrt(dx*dx + dy*dy)

    return dist
```

```python
def getO2(analog,temp,depths):
    DO_A = -4.262146e+01;
    DO_B = +1.426020e+02;
    DO_C = -3.346404e-01;
    DO_D = +1.110000e-02;
    DO_E = +4.500000e-03;
    DO_F = +0.000000e+00;
    DO_G = +0.000000e+00;
    DO_H = +1.000000e+00;
    
    M,N=np.shape(temp)
    Pprime = DO_A / (1 + DO_D *(temp - 25.))+DO_B / ((analog - DO_F) * (1. + DO_D *(temp - 25.)) + DO_C + DO_F);
    DO = (DO_G + DO_H * Pprime) * (1.+DO_E * np.tile(reshape(depths,(M,1)),(1,N)));
    return DO
```

```python
# objective analysis on a few depths?

sigma = 0.5;
def interpnaive((x,y),z,(xi,yi),sigma):
    nx=shape(xi)[0]
    ny=shape(yi)[0]
    N = shape(x)[0]
   
    zi = zeros((ny,nx))
    zactual = zeros((ny,nx))
    sumb=zeros((ny,nx))
    rall=zeros((ny,nx))
    for indx in range(nx):
        for indy in range(ny):
            r2 = (x-xi[indx])**2 + (y-yi[indy])**2
            b = exp(-r2/sigma**2)
            sumb[indy,indx]=sum(b)
            b = b/sum(b)
            zi[indy,indx]=sum(multiply(b,z))
            zactual[indy,indx]=xi[indx]*(exp(-xi[indx]**2 - yi[indy]**2))
            rall[indy,indx]=min(sqrt(r2))
    return zi,zactual,sumb,rall

def interpnaiveline((x,y),z,(xi,yi),sigma):
    nx=shape(xi)[0]
    ny=shape(yi)[0]
    N = shape(x)[0]
   
    zi = zeros((nx))
    zactual = zeros((nx))
    sumb=zeros((nx))
    rall=zeros((nx))
    for indx in range(nx):

        r2 = (x-xi[indx])**2 + (y-yi[indx])**2
        b = exp(-r2/sigma**2)
        sumb[indx]=sum(b)
        b = b/sum(b)
        zi[indx]=sum(multiply(b,z))
        zactual[indx]=xi[indx]*(exp(-xi[indx]**2 - yi[indx]**2))
        rall[indx]=min(sqrt(r2))
    return zi,zactual,sumb,rall

def interpbarnes((x,y),z,(xi,yi),sigma,Nit):
    nx=shape(xi)[0]
    ny=shape(yi)[0]
    N = shape(x)[0]
    zi = zeros((nx,ny))
    zactual = zeros((nx,ny))
    E = 0.*x
    zk = 0.*x

    zk2 = 0.*x
    for indx in range(nx):
        for indy in range(ny):
            r2 = (x-xi[indx])**2 + (y-yi[indy])**2
            b = exp(-r2/sigma**2)
            b = b/sum(b)
            zi[indy,indx]=sum(multiply(b,z))
    for k in range(0,N):
        r = (x-x[k])**2 + (y-y[k])**2
        b = exp(-r/sigma**2)
        b = b/sum(b)
        zk[k]=sum(multiply(b,z))
    E = z-zk[k]
   # plot(E)
   # print E
    # second pass, correcting with E.  
    for indx in range(nx):
        for indy in range(ny):
            r2 = (x-xi[indx])**2 + (y-yi[indy])**2
            b = exp(-r2/sigma**2)
            b = b/sum(b)
            zi[indy,indx]=zi[indy,indx]+sum(multiply(b,E))                
            zactual[indy,indx]=xi[indx]*(exp(-xi[indx]**2 - yi[indy]**2))
    # third pass
    for k in range(0,N):
        r = (x-x[k])**2 + (y-y[k])**2
        b = exp(-r/sigma**2)
        b = b/sum(b)
        zk2[k]=sum(multiply(b,zk))
    E = z-zk2
    for indx in range(nx):
        for indy in range(ny):
            r2 = (x-xi[indx])**2 + (y-yi[indy])**2
            b = exp(-r2/sigma**2)
            b = b/sum(b)
            zi[indy,indx]=zi[indy,indx]+sum(multiply(b,E))                
            zactual[indy,indx]=xi[indx]*(exp(-xi[indx]**2 - yi[indy]**2))
    zactual=zactual-zactual.mean()
    return zi,zactual

lon0=-25.6
lat0=48.4
sl['pden']=sl['pden'][0:64]
ind = where(sl['pden']>26.4)[0][0]
print ind
z=sl['spice'][ind,:]
good = where(~isnan(z))[0]

x= (sl['longitude'][good]-lon0)*cos(lat0*pi/180)*111.32
y= (sl['latitude'][good]-lat0)*111.32
z=z[good]
xi=arange(-60,60.,1.)
yi=arange(-60,60.,1.)
sigma=2.

zi,zactual,sumb,r = interpnaive((x, y), z, (xi, yi),sigma)
zi=ma.masked_where(r>5.,zi)
```

```python
def getO2(analog,temp,depths):
    DO_A = -4.262146e+01;
    DO_B = +1.426020e+02;
    DO_C = -3.346404e-01;
    DO_D = +1.110000e-02;
    DO_E = +4.500000e-03;
    DO_F = +0.000000e+00;
    DO_G = +0.000000e+00;
    DO_H = +1.000000e+00;
    
    M,N=np.shape(temp)
    Pprime = DO_A / (1 + DO_D *(temp - 25.))+DO_B / ((analog - DO_F) * (1. + DO_D *(temp - 25.)) + DO_C + DO_F);
    DO = (DO_G + DO_H * Pprime) * (1.+DO_E * np.tile(reshape(depths,(M,1)),(1,N)));
    return DO

cgrid['O2']=getO2(cgrid['analog'],cgrid['temp'],cgrid['depths'])
```
