```python
def bindata1d(binx,x,X):

    N = size(binx,0)-1
    
    #print shape(x)
    #print shape(X)
    
    meanX = zeros(N)
    varX = zeros(N)
    nX = zeros(N)
    
    #print 'Binx %d' % shape(binx)
    #print 'x %d ' % shape(x)
    #print shape(arange(0,N,1.0))
    
    indx = floor(interp(x,binx,arange(0,N+1,1.0)))
    indx[indx<0]=0
    indx[indx>N-1]=N-1
    #print 'Len x %d' %  len(x)
    for ind in range(len(x)):
        #print ind
        #print indx[ind]
        #print shape(X)
        meanX[indx[ind]]+=X[ind]
        nX[indx[ind]]+=1
    nX[nX==0.0]=NaN;
    meanX = meanX/nX;

    for ind in range(len(x)):
        varX[indx[ind]]+=(meanX[indx[ind]]-X[ind])**2
    varX=varX/nX
    # mask
    meanX=ma.masked_where(isnan(nX),meanX)
    nX=ma.masked_where(isnan(nX),nX)
    varX=ma.masked_where(isnan(nX),varX)
    
    
    return meanX,varX,nX
```

```python
def gridctd(fname):
    import numpy as np
   

    data=np.genfromtxt(fname,comments='#',autostrip=True,skip_header=320,invalid_raise=True) 
    
    togrid = ['cond', 'sal', 'temp', 'v2', 'v3', 'trans', 'atten', 'O2sat', 'O2umolKg', 'sigmat']
    rows = [3,        5,     7,      11,  12,       14,    15,    17,       19,  20]
    N = 24*20
    p=data[:,2]
    dp = convolve(diff(p),ones(N)/N,mode='same')
    down = where(dp>0.01)[0]
    pgrid = arange(0.,2000.,1.)

    ctdgrid=dict()
    ctdgrid['depths']=pgrid[:-1]+0.5
    for row,gridname in zip(rows,togrid):
        y = data[:,row]
        y = squeeze(y[down])
        ctdgrid[gridname],v,vv=bindata1d(pgrid,p[down],y)
    
    # parse the lat,lon and time....
    with open(fname,'r') as fin:
        for line in fin:
            if "NMEA Latitude =" in line:
                a=line.split()
                ctdgrid['lat']=float(a[4])+float(a[5])/60.
                print ctdgrid['lat']
                if a[6]=='S':
                    ctdgrid['lat']=-ctdgrid['lat']
            if "NMEA Longitude =" in line:
                a=line.split()
                ctdgrid['lon']=float(a[4])+float(a[5])/60.
                if a[6]=='W':
                    ctdgrid['lon']=-ctdgrid['lon']
                print ctdgrid['lon']
            if "NMEA UTC (Time) =" in line:
                print line
        
    
    return ctdgrid
```

```python
import glob
fnames=glob.glob('../FK009A/Oceanography/CTD/*.cnv')
print shape(fnames)
print fnames[0]
ctdgrid=dict()
n=0
for fname in fnames:
    print fname
    cgrid=gridctd(fname)
    
    keys = cgrid.keys()
   
    for key in keys:
        if (n==0):
            N = shape(cgrid['temp'])[0]
            if shape(cgrid[key]):
                ctdgrid[key] = np.reshape(cgrid[key],(N,1))   
            else:
                ctdgrid[key]=cgrid[key]
        else:
            if shape(cgrid[key]):
                ctdgrid[key] = np.hstack((ctdgrid[key],np.reshape(cgrid[key],(N,1))))
            else:
                ctdgrid[key] = np.hstack((ctdgrid[key],cgrid[key]))
    n=n+1
            
```

```python
from jmkfigure import *
import cPickle as pickle
import scipy.io as sio
def plotTopo(clab=False):
    topo=sio.loadmat('../topo/ViTopo.mat')
    indx = where((squeeze(topo['topolon'])>-127.) & (squeeze(topo['topolon'])<-124.))[0]
    indy = where(squeeze(topo['topolat']>47.5) & squeeze(topo['topolat']<49.2))[0]
    
    matplotlib.rcParams['contour.negative_linestyle'] = 'solid'
    x=squeeze(topo['topolon'])[indx]
    y = squeeze(topo['topolat'])[indy]
    Z=topo['topodepth'][indy,:][:,indx]
    contourf(x,y,Z,
            arange(-3000.,00.,200.),cmap=cm.ocean)
    clim([-4000,900./(900-300)*4300.-4000.])
    contourf(x,y,Z,
            arange(-300.,0,25),cmap=cm.ocean)
    clim([-900,0.])
#    clim([-3500,0.])
    c=contour(x,y,-Z,
            [1000.,200.,100.],colors='0.5',linestyles='-',linewidth=1.5)
    if clab:
        xlim([-126.-25./60.,-124.-40./60.])
        ylim([47.8,48.8])
        gca().set_aspect(1./cos(48.5*pi/180.))
        manual_locations = [(-125.25, 48.65), (-126.1, 48.45),(-125.7,47.9)]
        clabel(c,fontsize=7,fmt='%d',manual=manual_locations)
    contourf(x,y,Z,
            [0.,1000000.],colors='darkgreen',linewidth=3)
```
